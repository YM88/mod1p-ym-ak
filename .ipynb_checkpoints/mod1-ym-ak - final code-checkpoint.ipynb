###  **title.basics.csv.gz** - *dataframe used for **genres**, **original_title**, and the **start_year** information*
Contains the following information for titles:
- tconst (string) - alphanumeric unique identifier of the title
- titleType (string) – the type/format of the title (e.g. movie, short, tvseries, tvepisode, video, etc)
- primaryTitle (string) – the more popular title / the title used by the filmmakers on promotional materials at the point of release
- originalTitle (string) - original title, in the original language
- isAdult (boolean) - 0: non-adult title; 1: adult title
- startYear (YYYY) – represents the release year of a title. In the case of TV Series, it is the series start year

- runtimeMinutes – primary runtime of the title, in minutes
- genres (string array) – includes up to three genres associated with the title

````python
pd.read_csv('data/imdb.title.basics.csv.gz').head()
````

````python
# creating the imdb_title_basics_df through pandas
imdb_title_basics_df = pd.read_csv('data/imdb.title.basics.csv.gz',
                                   usecols=['tconst',
                                            'original_title',
                                            'start_year',
                                            'genres'])
imdb_title_basics_df.head()
imdb_title_basics_df.info()
````

````python
# creating a list of all genres mentioned in the title_basics_df
all_genres = set()
genre_series = imdb_title_basics_df.genres

for row in genre_series:
    try:
        for genre in row.split(','):
            all_genres.add(genre)
    except:
        continue
all_genres = list(all_genres)
````

````python
# creating separate columns for each genre in the title_basics_df
for genre in all_genres:
    imdb_title_basics_df[genre] = imdb_title_basics_df.genres.apply(
        lambda x: genre in str(x).split(','))

imdb_title_basics_df.columns
imdb_title_basics_df.head()
````

## 5. [The Numbers](https://www.the-numbers.com/movie/budgets/all) Data - _**release_date**, **movie**, **production_budget**, and **worldwide_gross** values to be used in the EDA._

````python
pd.read_csv('data/tn.movie_budgets.csv.gz').head()
tn_movie_budgets_df = pd.read_csv('data/tn.movie_budgets.csv.gz')
````

````python
def convert_to_numeric(some_object):
    '''This function converts formatted numbers 
     with $ sign and thousand separators to integers.'''
    return int(some_object.lstrip('$').replace(',', ''))


# convert formatted numbers to numeric values
tn_movie_budgets_df.iloc[:, 3:] = tn_movie_budgets_df.iloc[:, 3:].applymap(
    convert_to_numeric)
# remove id column
tn_movie_budgets_df.drop('id', inplace=True, axis=1)
# keep only the year from the release_date column
tn_movie_budgets_df['year'] = tn_movie_budgets_df.release_date.apply(
    lambda x: x[-4:]).astype(int)
tn_movie_budgets_df.drop('release_date', inplace=True, axis=1)
tn_movie_budgets_df.head()
tn_movie_budgets_df.info()
````

# II. EDA

## 1. Merging Dataframes of imdb title basics and The Numbers for financial analysis per genre

Steps include:

- Joining the dataframes and creating the `genre_financial_df`
- Filtering for 2010-2018 years
- Removing NaN in genre and worldwide values 
- Removing duplicate in movie name values
- creating new descriptive values:

`1. net income = **worldwide gross** - **production budget**`


`2. net income ratio = **net income** / **production budget**`

- creating a new dataframe for descriptive analysis on genres `describe_genres`


````python
genre_financial_df = pd.merge(imdb_title_basics_df, tn_movie_budgets_df,
                              left_on='original_title',
                              right_on='movie',
                              how='inner')
````

````python
genre_financial_df.info()
````

````python
genre_financial_df = genre_financial_df.loc[(genre_financial_df.year > 2009) &  # filtering out movies for 2010-2018
                                            (genre_financial_df.year < 2019)]
````

````python
genre_financial_df.head()
genre_financial_df.info()
````

````python
# using the year and the start year to filter out duplicate movies in different years and use one
genre_financial_df = genre_financial_df.loc[genre_financial_df.year ==
                                            genre_financial_df.start_year]
````

````python
genre_financial_df.genres.isna().sum()  # finding the none values
````

````python
genre_financial_df.dropna(subset=['genres'], inplace=True)
````

````python
genre_financial_df = genre_financial_df.loc[(
    genre_financial_df.worldwide_gross != 0)]
````

````python
# looking for duplicate values on movies that have the same year and movie names/titles
display(genre_financial_df[genre_financial_df.duplicated(
    subset=['original_title', 'year'], keep=False)])
# setting panda code settin gto display all row values without truncation
pd.options.display.max_rows = 77
````

````python
# manually finding which indices to remove per the numbers  and the imdb websites
duplicate_indices = [210, 216, 225, 276, 456, 505, 540, 700, 714, 781,
                     805, 1012, 1013, 1079, 1244, 1261, 1274, 1390, 1499,
                     1500, 1532, 1704, 1732, 1800, 1833, 1904, 2055, 2061,
                     2165, 2230, 2442, 2443, 2563, 2657, 2662, 2664, 2823,
                     2831, 2925, 3037, 3176]
# removing the duplicates
genre_financial_df.drop(index=duplicate_indices, inplace=True)
````

````python
# creating the net_income column
genre_financial_df['net_income'] = genre_financial_df['worldwide_gross'] - \
    genre_financial_df['production_budget']
````

````python
# creating the net_income_ratio
genre_financial_df['net_income_ratio'] = genre_financial_df['net_income'] / \
    genre_financial_df['production_budget']
````

````python
genre_financial_df.head()
````

````python
# all_genres list were created in part I.2. title_basics_df
print(all_genres)
````

````python
# creating the describe_genres_df for aggregated information on genres and budget
describe_genres = []

for genre in all_genres:
    # We will add financial data in case of every genre, collected in the following list:
    describe_genre = []
    describe_genre.append(genre)
    # add the number of the films listed in that
    describe_genre.append(
        genre_financial_df.loc[genre_financial_df[genre], 'production_budget'].count())
    # add worldwide gross mean
    describe_genre.append(
        genre_financial_df.loc[genre_financial_df[genre], 'worldwide_gross'].mean())

    # add worldwide gross median
    describe_genre.append(
        genre_financial_df.loc[genre_financial_df[genre], 'worldwide_gross'].median())

    # add production budget mean
    describe_genre.append(
        genre_financial_df.loc[genre_financial_df[genre], 'production_budget'].mean())
    # add production budget median
    describe_genre.append(
        genre_financial_df.loc[genre_financial_df[genre], 'production_budget'].median())
    # add net_income mean
    describe_genre.append(
        genre_financial_df.loc[genre_financial_df[genre], 'net_income'].mean())
    # add production budget median
    describe_genre.append(
        genre_financial_df.loc[genre_financial_df[genre], 'net_income'].median())
    # add net income ratio mean
    describe_genre.append(
        genre_financial_df.loc[genre_financial_df[genre], 'net_income_ratio'].mean())
    # add production budget median
    describe_genre.append(
        genre_financial_df.loc[genre_financial_df[genre], 'net_income_ratio'].median())
    describe_genres.append(describe_genre)

describe_genres_df = pd.DataFrame(describe_genres, columns=['genre', 'count', 'wordlwide_gross_mean',
                                                            'worldwide_gross_median', 'production_budget_mean',
                                                            'production_budget_median', 'net_income_mean',
                                                            'net_income_median', 'net_income_ratio_mean',
                                                            'net_income_ratio_median'])
````

````python
# checking the output of the describe_genres_df
describe_genres_df.head()
````

````python
# removed genres with 0 entries
describe_genres_df = describe_genres_df.loc[describe_genres_df['count'] > 0]
# data will be in $ million, ratio in %
describe_genres_df.iloc[:, 2:8] = describe_genres_df.iloc[:, 2:8]/1000000
describe_genres_df.iloc[:, 8:] = describe_genres_df.iloc[:, 8:]*100
describe_genres_df.iloc[:, 2:] = describe_genres_df.iloc[:, 2:].applymap(
    lambda x: float(f'{x:.2f}'))
describe_genres_df.head()
````

````python
# sorting the df from the highest to lowest net_median_income
describe_genres_df.sort_values(
    by='net_income_median', inplace=True, ascending=True)
describe_genres_df.head()
````

## 2. Using financial data from EDA.1 part and merge with**bom_movie_gross** data
Steps include: 

- Removing NaN values from Studios
- Creating a new studio_financial _df dataframe by merging bom_movie_gross_df and genre_financial_df
- Renaming the studio name values
- Creating a subset df with the top 4 genres in highest net income medians
- Creating a subset df with the 2 top performing studios in the low budget

````python
# removing rows with null values in studio column
bom_movie_gross_df.dropna(subset=['studio'], inplace=True)
````

````python
# removing rows null value rows in domestic gross column
# subset created just to analyze the studio's income
bom_movie_gross_df_subset = bom_movie_gross_df.dropna(
    subset=['domestic_gross'])
bom_movie_gross_df_subset.domestic_gross.astype(int)
````

````python
# creating the studio_financial_df
studio_financial_df = pd.merge(bom_movie_gross_df, genre_financial_df,
                               left_on='title', right_on='original_title', how='inner')
# create a subset with the targeted genres
focused_subset_df = studio_financial_df.loc[(studio_financial_df['Animation'] == True)
                                            | (studio_financial_df['Adventure'] == True)
                                            | (studio_financial_df['Sci-Fi'] == True)
                                            | (studio_financial_df['Action'] == True)]
# checking the new df
focused_subset_df.head()
````

````python
# renaming the studio values
studio_names = {'Uni.': 'Universal', 'P/DW': 'P. DreamWorks', 'BV': 'Buena Vista',
                'ORF': 'Open Road Films', 'Orch.': 'Orchard', 'FoxS': 'Fox Searchlight',
                'Magn.': 'Magnolia', 'WB (NL)': 'Warner Bros', 'W/Dim.': 'Dimension',
                'Rela.': 'Relativity', 'WB': 'Warner Bros', 'Par.': 'Paramount',
                'SGem': 'Screen Gems'}


def rename_studio(orig_name):
    try:
        new_name = studio_names[orig_name]
        return new_name
    except:
        return orig_name


focused_subset_df.studio = focused_subset_df.studio.apply(rename_studio)
# convert net income ratio to % value
focused_subset_df.net_income_ratio = focused_subset_df.net_income_ratio * 100
focused_subset_df.head()
````

````python
anim_studios = focused_subset_df.loc[focused_subset_df['Animation'] == True].groupby(
    'studio')['net_income_ratio'].median().sort_values(ascending=False)
adv_studios = focused_subset_df.loc[focused_subset_df['Adventure'] == True].groupby(
    'studio')['net_income_ratio'].median().sort_values(ascending=False)
scifi_studios = focused_subset_df.loc[focused_subset_df['Sci-Fi'] == True].groupby(
    'studio')['net_income_ratio'].median().sort_values(ascending=False)
action_studios = focused_subset_df.loc[focused_subset_df['Action'] == True].groupby(
    'studio')['net_income_ratio'].median().sort_values(ascending=False)

anim_studios = anim_studios.map(lambda x: f'{x:.2f}').astype(float)
````

````python
focused_subset2_df = studio_financial_df.loc[(studio_financial_df['Horror'] == True)
                                             | (studio_financial_df['Mystery'] == True)]


focused_subset2_df.studio = focused_subset2_df.studio.apply(rename_studio)
# convert net income ratio to % value
focused_subset2_df.net_income_ratio = focused_subset2_df.net_income_ratio * 100


focused_subset2_df.head(2)
horror_studios = focused_subset2_df.loc[focused_subset2_df['Horror'] == True].groupby(
    'studio')['net_income_ratio'].median().sort_values(ascending=False)
mystery_studios = focused_subset2_df.loc[focused_subset2_df['Mystery'] == True].groupby(
    'studio')['net_income_ratio'].median().sort_values(ascending=False)
````

# III. Visualizations

## 1. Comparison of Medians in Worldwide Revenues and Production Budgets

````python
sns.set(style="whitegrid")
df = describe_genres_df.set_index('genre')
df = df.sort_values(by='worldwide_gross_median')
df = df.loc[:, ['worldwide_gross_median', 'production_budget_median']]
fig1 = plt.figure(figsize=(20, 14))
y_pos = np.arange(len(df))
plt.barh(y_pos+0.2, df.worldwide_gross_median, height=0.4,
         color='red', alpha=0.7, label='worldwide gross income')
plt.barh(y_pos-0.2, df.production_budget_median,
         color='pink', height=0.4, label='production budget')
plt.yticks(y_pos, df.index)
plt.title('Median Worldwide Revenue and Budget in $M by Movie Genres (2010-2018)\n', fontsize=30)
plt.gca().xaxis.set_tick_params(labelsize=20)
plt.gca().yaxis.set_tick_params(labelsize=20)
plt.gca().legend(loc=4, fontsize=20)
fig1.savefig('fig_1_Revenue_budget.jpg')
plt.show()
````

## 2. Median of Worldwide Net Income in Different Movie Genres

````python
sns.set(style="whitegrid")
df = describe_genres_df.set_index('genre')
df = df.sort_values(by='net_income_median')
df = df.loc[:, ['net_income_median', 'production_budget_median']]
fig2 = plt.figure(figsize=(20, 20))
y_pos = np.arange(len(df))
plt.barh(y_pos, df.net_income_median, alpha=1, label='net income')
plt.yticks(y_pos, df.index)
plt.title(
    'Median Worldwide Net Income in $M by Movie Genres (2010-2018)\n', fontsize=30)
plt.gca().xaxis.set_tick_params(labelsize=20)
plt.gca().yaxis.set_tick_params(labelsize=20)
fig2.savefig('fig_2_Net_income.jpg')
plt.show()
````

## 3. Comparison Between Median Net Income Ratio & Budget Values per Genres

````python
sns.set(style="whitegrid")
df = describe_genres_df.set_index('genre')
df = df.sort_values(by='net_income_median')
df = df.loc[:, ['net_income_ratio_median', 'production_budget_median']]
plt.figure(figsize=(20, 20))
c = '#358188'
fig3, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(16, 8))
ax1.barh(describe_genres_df.genre, describe_genres_df.net_income_ratio_median,
         color=c, label='net income ratio')
ax2.barh(describe_genres_df.genre,
         describe_genres_df.production_budget_median, label='production budget')
ax1.yaxis.set_tick_params(labelsize=16)
ax1.set_title('Median Net income ratio (%) by genres', fontsize=20)
ax2.set_title('Median Budget $M by genres', fontsize=20)
fig3.savefig('fig_3_Net_incom_ratio_budget.jpg')
plt.show()
````

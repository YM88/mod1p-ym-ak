# 2. IMDB Data

Data gathered from [IMDB](https://www.imdb.com/interfaces/)

## name values like actor/director to be taken out from this dataset

**name.basics.tsv.gz** – Contains the following information for names:
- nconst (string) - alphanumeric unique identifier of the name/person
- primaryName (string)– name by which the person is most often credited
- birthYear – in YYYY format
- deathYear – in YYYY format if applicable, else '\N'
- primaryProfession (array of strings)– the top-3 professions of the person
- knownForTitles (array of tconsts) – titles the person is known for

```python
names_df = pd.read_csv('data/imdb.name.basics.csv.gz')

names_df.head()
```

**title.akas.tsv.gz** - Contains the following information for titles:

- titleId (string) - a tconst, an alphanumeric unique identifier of the title
- ordering (integer) – a number to uniquely identify rows for a given titleId
- title (string) – the localized title
- region (string) - the region for this version of the title
- language (string) - the language of the title
- types (array) - Enumerated set of attributes for this alternative title. One or more of the following: "alternative", "dvd", "festival", "tv", "video", "working", "original", "imdbDisplay". New values may be added in the future without warning
- attributes (array) - Additional terms to describe this alternative title, not enumerated
- isOriginalTitle (boolean) – 0: not original title; 1: original title

```python
#checking the dataframe, won't be using this one
pd.read_csv('data/imdb.title.akas.csv.gz').head()
```

## Genre & title name info to be merged from this dataframe
**title.basics.tsv.gz** - Contains the following information for titles:
- tconst (string) - alphanumeric unique identifier of the title
- titleType (string) – the type/format of the title (e.g. movie, short, tvseries, tvepisode, video, etc)
- primaryTitle (string) – the more popular title / the title used by the filmmakers on promotional materials at the point of release
- originalTitle (string) - original title, in the original language
- isAdult (boolean) - 0: non-adult title; 1: adult title
- startYear (YYYY) – represents the release year of a title. In the case of TV Series, it is the series start year
- endYear (YYYY) – TV Series end year. ‘\N’ for all other title types
- runtimeMinutes – primary runtime of the title, in minutes
- genres (string array) – includes up to three genres associated with the title

```python
pd.read_csv('data/imdb.title.basics.csv.gz').head()
```

```python
imdb_title_basics_df = pd.read_csv('data/imdb.title.basics.csv.gz', 
                                   usecols = ['tconst', 
                                              'original_title', 
                                              'start_year', 
                                             'genres'])
imdb_title_basics_df.head()
imdb_title_basics_df.info()
```

```python
#to get a list of all mentioned genres
all_genres = set()
genre_series = imdb_title_basics_df.genres

for row in genre_series:
    try:
        for genre in row.split(','):
            all_genres.add(genre)
    except:
        continue
all_genres = list(all_genres)

#created separate columns for the genres
for genre in all_genres:
    imdb_title_basics_df[genre] = imdb_title_basics_df.genres.apply(lambda x: genre in str(x).split(','))

imdb_title_basics_df.columns
imdb_title_basics_df.head()
```

```python
def convert_to_numeric(some_object):
    '''This function converts formatted numbers 
     with $ sign and thousand separaters to integers.'''
    return int(some_object.lstrip('$').replace(',',''))

#convert formatted numbers to numeric values
tn_movie_budgets_df.iloc[:, 3:] = tn_movie_budgets_df.iloc[:, 3:].applymap(convert_to_numeric)
#remove id column
tn_movie_budgets_df.drop('id', inplace=True, axis=1)
#keep only the year from the release_date column
tn_movie_budgets_df['year'] = tn_movie_budgets_df.release_date.apply(lambda x: x[-4:]).astype(int)
tn_movie_budgets_df.drop('release_date', inplace=True, axis=1)
tn_movie_budgets_df.head()
tn_movie_budgets_df.info()
```

```python
genre_financial_df = pd.merge(imdb_title_basics_df, tn_movie_budgets_df, 
         left_on = 'original_title', 
        right_on = 'movie', 
        how = 'inner')
```

```python
# aggregated information on genres and budget
describe_genres=[]

for genre in all_genres:
    #We will add financial data in case of every genre, collected in the following list:
    describe_genre = []
    describe_genre.append(genre)
    #add the number of the films listed in that 
    describe_genre.append(genre_financial_df.loc[genre_financial_df[genre], 'production_budget'].count())
    #add worldwide gross mean
    describe_genre.append(genre_financial_df.loc[genre_financial_df[genre], 'worldwide_gross'].mean())

    #add worldwide gross median
    describe_genre.append(genre_financial_df.loc[genre_financial_df[genre], 'worldwide_gross'].median())
    
    
    
    #add production budget mean
    describe_genre.append(genre_financial_df.loc[genre_financial_df[genre], 'production_budget'].mean())
    #add production budget median
    describe_genre.append(genre_financial_df.loc[genre_financial_df[genre], 'production_budget'].median())
    #add net_income mean
    describe_genre.append(genre_financial_df.loc[genre_financial_df[genre], 'net_income'].mean())
    #add production budget median
    describe_genre.append(genre_financial_df.loc[genre_financial_df[genre], 'net_income'].median())
    #add net income ratio mean
    describe_genre.append(genre_financial_df.loc[genre_financial_df[genre], 'net_income_ratio'].mean())
    #add production budget median
    describe_genre.append(genre_financial_df.loc[genre_financial_df[genre], 'net_income_ratio'].median())
    describe_genres.append(describe_genre)
    
describe_genres_df = pd.DataFrame(describe_genres, columns=['genre', 'count', 'wordlwide_gross_mean', 
                                                           'worldwide_gross_median', 'production_budget_mean',
                                                           'production_budget_median', 'net_income_mean',
                                                           'net_income_median', 'net_income_ratio_mean', 
                                                           'net_income_ratio_median'])
len(describe_genres_df)
```

```python
#removed genres with 0 entries
describe_genres_df = describe_genres_df.loc[describe_genres_df['count']>0]
# data will be in $ million, ratio in %
describe_genres_df.iloc[:, 2:8] = describe_genres_df.iloc[:, 2:8]/1000000
describe_genres_df.iloc[:, 8:] = describe_genres_df.iloc[:, 8:]*100
describe_genres_df.iloc[:, 2:] = describe_genres_df.iloc[:, 2:].applymap(lambda x: float(f'{x:.2f}'))
describe_genres_df.head()
```

```python
describe_genres_df.sort_values(by='net_income_median', inplace=True, ascending=True)
describe_genres_df.head()
```

```python
#bar chart on genres and budget information

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(16,8))

ax1.barh(describe_genres_df.genre, describe_genres_df.net_income_median)
ax2.barh(describe_genres_df.genre, describe_genres_df.production_budget_median, color='darkblue')

# Set a title for the presentation: 'Net income and budget by movie genres (2009-2018)'

ax1.set_title('Net income median in million dollar')
ax2.set_title('Budget median in million dollar')

plt.show()
```

### Studio and budget information

```python
# merge with bom

studio_financial_df = pd.merge(bom_movie_gross_df, genre_financial_df, left_on = 'title', right_on = 'original_title', how='inner')
```

```python
# create a subset with the targeted genres
focused_subset_df = studio_financial_df.loc[(studio_financial_df['Animation']==True) 
                                       | (studio_financial_df['Adventure']==True)
                                       | (studio_financial_df['Sci-Fi']==True)
                                       | (studio_financial_df['Action']==True)]
focused_subset_df.head(2)
```

```python
# rename studios
studio_names = {'Uni.': 'Universal', 'P/DW': 'P. DreamWorks', 'BV': 'Buena Vista', 
                'ORF': 'Open Road Films', 'Orch.': 'Orchard', 'FoxS':'Fox Searchlight',
               'Magn.': 'Magnolia', 'WB (NL)': 'Warner Bros', 'W/Dim.' : 'Dimension',
               'Rela.' : 'Relativity', 'WB':'Warner Bros', 'Par.' : 'Paramount',
               'SGem': 'Screen Gems'}

def rename_studio(orig_name):
    try:
        new_name = studio_names[orig_name]
        return new_name
    except:
        return orig_name

focused_subset_df.studio = focused_subset_df.studio.apply(rename_studio)
# convert net income ratio to % value
focused_subset_df.net_income_ratio = focused_subset_df.net_income_ratio * 100
focused_subset_df.head()
```

```python

# focused_subset_df.studio.value_counts()
anim_studios = focused_subset_df.loc[focused_subset_df['Animation'] == True].groupby('studio')['net_income_ratio'].median().sort_values(ascending=False)
adv_studios = focused_subset_df.loc[focused_subset_df['Adventure'] == True].groupby('studio')['net_income_ratio'].median().sort_values(ascending=False)
scifi_studios = focused_subset_df.loc[focused_subset_df['Sci-Fi'] == True].groupby('studio')['net_income_ratio'].median().sort_values(ascending=False)
action_studios = focused_subset_df.loc[focused_subset_df['Action'] == True].groupby('studio')['net_income_ratio'].median().sort_values(ascending=False)
horror_studios = focused_subset_df.loc[focused_subset_df['Horror'] == True].groupby('studio')['net_income_ratio'].median().sort_values(ascending=False)
mystery_studios = focused_subset_df.loc[focused_subset_df['Mystery'] == True].groupby('studio')['net_income_ratio'].median().sort_values(ascending=False)
anim_studios = anim_studios.map(lambda x: f'{x:.2f}').astype(float)
```

```python
# visualize the studios performance by genres
sns.set(style="whitegrid")
# pallete ="GnBu_d"
c = '#358188'

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16,12))

fig.subplots_adjust(hspace = 0.8)

plt.suptitle("Net income ratio (%) by studios for the top four genres\n\n\n\n", fontsize = 25)

sns.barplot(list(anim_studios[:6].index), anim_studios[:6], color = c, ax = ax1)
ax1.set_title('Animation genre\n', fontsize = 16)
ax1.set_ylabel('')
ax1.xaxis.set_tick_params(labelsize=15)
for item in ax1.get_xticklabels():
    item.set_rotation(45)

sns.barplot(list(adv_studios[:6].index), adv_studios[:6],  color = c,ax = ax2)
ax2.set_title('Adventure genre\n', fontsize = 16)
ax2.set_ylabel('')
ax2.xaxis.set_tick_params(labelsize=15)
for item in ax2.get_xticklabels():
    item.set_rotation(45)

sns.barplot(list(scifi_studios[:6].index), scifi_studios[:6], color = c,ax = ax3)
ax3.set_title('\nSci-Fi genre\n', fontsize = 16)
ax3.set_ylabel('')
ax3.xaxis.set_tick_params(labelsize=15)
for item in ax3.get_xticklabels():
    item.set_rotation(45)

sns.barplot(list(action_studios[:6].index), action_studios[:6],  color = c, ax = ax4)
ax4.set_title('\nAction\n', fontsize = 16)
ax4.set_ylabel('')
ax4.xaxis.set_tick_params(labelsize=15)
for item in ax4.get_xticklabels():
    item.set_rotation(45)

fig.savefig('fig_4_Studios.jpg')

plt.show();
```
